```python
# Import functions to read, show and save images
from skimage.io import imread,imshow,imsave
# Functions to work with matrices
import numpy as np
# Functions to convert volor images from rgb to hsv and back
from skimage.color import rgb2hsv, hsv2rgb
# Function to calculate entropy map
from skimage.filters.rank import entropy
%matplotlib inline
```

```python
g = imread('camera.jpg')
imshow(g)
imsave('camera_copy.jpg', g)
#Check if file 'camera-copy.jpg' exists in directory
import os.path
file_exists = os.path.exists('camera_copy.jpg')
print(file_exists)
```

```python
import numpy as np

## -- Your code here -- ##
print('Our image is stored as a matrix:')
print(g)
print('Shape of matrix:')
print(g.shape)
print('Max and min value of matrix:')
print(g.max(), g.min())
print('Grayscale image type')
print(g.dtype)
```

```python
g_cropped = g[100:200,200:300]
imshow(g_cropped)
```

```python
scanner = np.ones((15, 15))
im_entropy_map = entropy(g, scanner)
imshow(im_entropy_map)
```
